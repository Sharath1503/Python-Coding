## Data Processing function
- map() - same operation applied on all elements
- filter() - select elements based on condition

```python
map(int, ["10", "20", "30"])
```

```python
# simple extraction from map using loop
for num in map(int, ["10", "20", "30"]):
    print(num)
```

```python
for ascii_code in map(ord, ['A','b','c','d']):
    print(ascii_code, end = ' ')
```

```python
list(map(ord, ['A','b','c','d']))
```

```python
list(map(round, [1.5, 2.3, 3.6, 4.9]))
```

```python
tuple(map(round, [1.5, 2.3, 3.6, 4.9]))
```

```python
set(map(round, [1.5, 2.3, 3.6, 4.9]))
```

```python
animals = ['Horse', 'Rabbit', 'Deer', 'Antelope', 'Giraffe', 'Lion', 'Tiger', 'Wolf', 'Kangaroo', 'Meercat', 'Wolverine']
out = list(map(lambda e: e[0], animals))
print(out)
```

```python
p = list(map(lambda e: e[:3], animals))
print(p)
```

```python
fruits = ['Apple','Banana','cherry', 'Water Melon', 'Orange', 'Dragon fruit', 'Berry', 'Cherry', 'Hazelnut', 'Grape', 'Guava']
print(len(animals)==len(fruits))
```

```python
# first 3 char of animal and last 3 char of fruit
list(map(lambda a,f: a[:3] + f[-3:], animals, fruits))
```

```python
x, y = [1,2,3] , [3,4,5]
z = list(map(lambda i,j: i+j, x, y))
print(x,y,z, sep="\n")
```

## filter()

```python
x = [1, -2, -3, 4, 5]
ans = filter(lambda i: i > 0, x)
list(ans)
```

```python
# retrieve names that do have 'a' in them
p = filter(lambda name: 'a' in name, animals)
list(p)
```

```python
# retrieve names that do not have 'a' in them
p = filter(lambda name: 'a' not in name, animals)
list(p)
```

```python
# filter names that have more than 5 chars
p = filter(lambda e: len(e) > 5, animals)
list(p)
```

```python
# a better example - get all the words file in downloads folder
import os
path = 'E:/BCA/Projects'
p = filter(lambda file: file.endswith('pdf'), os.listdir(path))
list(p)
```

```python
# list all files that has length greater than 2 chars
path = 'C:/Program Files'
p = list(filter(lambda item: len(item) > 2, os.listdir(path)))
p
```

```python
class Student:
    def __init__(self, name, marks=[]):
        self.name = name
        self.marks = marks
    
    def total_marks(Self): 
        return sum(Self.marks)
    
    def __gt__(self, other): 
        if isinstance(other, Student):
            return self.total_marks() > other.total_marks()
        elif isinstance(other, int):
            return self.total_marks() > other
    
    def __repr__(self):
        return f'Student({self.name}, {self.total_marks()})'
   
def is_passed(marks):
    return marks > 200

students = [
    Student('Zaid', [20,30,50,70,50]),
    Student('Kamil', [40,50,60,7,9]),
    Student('Ghouse', [20,34,45, 40,40]),
    Student('Aslam', [25,35,57,72,22]),
    Student('Aquib', [25,35,55,75,65])
]
list(filter(is_passed, students))
```
