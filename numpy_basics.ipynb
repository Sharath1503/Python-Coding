```python
import numpy as np
n = np.array([1,2,3,4])
print(n)
print(type(n))
```

```python
n2 = np.array([[1,2,3,], [4,5,6]])
print(n2)
print(type(n2))
```

```python
n3 = np.array([[[1,2], [5,6], [7,8]]])
print(n3)
print(type(n3))
```

```python
print(n.shape,n2.shape, n3.shape)
print(n.ndim, n2.ndim,n3.ndim)
print(n.size, n2.size)
```

```python
a = np.array([1,2,3])
b = np.array([3,4,5])
c = a + b
d = a-b
e = a/b
f = a*b
print(c,d,e,f, sep="\n")
```

```python
print(np.sin(a))
print(np.cos(b))
print(np.cos(c))
```

## Maths & Stats

```python
x = np.array([[2,4],[5,11],[3,4]])
print(x.mean())
print(x.mean(axis=0))
print(x.mean(axis=1))
```

```python
print(x.sum(axis=1))
print(x.max(axis=1))
print(x.min(axis=1))
```
