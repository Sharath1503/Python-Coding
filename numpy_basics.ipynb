```python
import numpy as np
n = np.array([1,2,3,4])
print(n)
print(type(n))
```

```python
n2 = np.array([[1,2,3,], [4,5,6]])
print(n2)
print(type(n2))
```

```python
n3 = np.array([[[1,2], [5,6], [7,8]]])
print(n3)
print(type(n3))
```
